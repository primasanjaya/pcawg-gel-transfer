```python
pd_allfeatures = pd.DataFrame()
for fo in range(1,11):
    permodel = all_bestmodel[fo-1]

    features = pd.read_csv(permodel + '/features.csv',index_col=0)
    listsample = pd.read_csv(permodel + '/listsample.csv',index_col=0)
    target = pd.read_csv(permodel + '/target.csv',index_col=0)

    M_list = []
    for i in range(1,25):
        M = 'M' + str(fo) + '_' + str(i)
        M_list.append(M)

    features.columns = M_list
    features['samples'] = listsample
    features['target'] = target
    
    features = features.reset_index(drop=True)
    if fo == 1:
        pd_allfeatures = features
    else:
        pd_allfeatures = pd_allfeatures.merge(features,left_on=['samples','target'], right_on = ['samples','target'],how='left')
```

```python
st = pd_allfeatures[['samples','target']]
```

```python
pd_allfeatures = pd_allfeatures.drop(['samples', 'target'],axis=1)
```

```python
pd_allfeatures_withmeta = pd.concat([pd_allfeatures, st], axis=1)
```

```python
pd_allfeatures_withmeta.to_csv('all_features_pcawg.csv')
```
